```python
from pathlib import Path
import pandas as pd

data_dir = Path("../data")
```

```python
csv_data = data_dir / 'data.csv'
```

```python
df = pd.read_csv(csv_data)
```

```python
df
```

```python
df.bist100.info
```

```python
def change_type(v: str | float):
    if isinstance(v, float):
        return v
    return v.strip('').strip(' ').replace(' ', '').strip().replace(',','.')
```

```python
df.bist100 = df.bist100.apply(change_type)
```

```python
df.bist100 = df.bist100.astype(float)
```

```python
df
```

```python
df.set_index(df.date)
```

```python
df.index
```

```python
df.index = pd.to_datetime(df.date)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
ror_daily = df.bist100.pct_change()
```

```python
ror_daily
```

```python
ror_weekly = df.bist100.resample('W').ffill().pct_change()
```

```python
ror_weekly
```

```python
ror_monthly = df.bist100.resample('M').ffill().pct_change()
```

```python
ror_monthly
```

```python
ror_bist100 = pd.DataFrame()
```

```python
ror_bist100
```

```python
ror_bist100['daily'] = ror_daily
```

```python
ror_bist100
```

```python
ror_bist100['weekly'] = ror_weekly
```
